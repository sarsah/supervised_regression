**Issues with In-Sample Data**
Because we used the same records to train and test, there could be issues of Overfitting.
To avoid that, the data needs to be split into training and test data before fitting the model

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,test_size=.3)
```

**Decsion Tree, Max-Nodes**
The number nodes on a tree can contribute significantly to the accuracy of a model. It can also impact overfitting and underfitting. Next, a function will be written to help tweak the model. Applying different max nodes until one with the best MAE is found

```python
#this function applies different maximum nodes to a DecisionTreeRegression algorithm
def get_mae(max_nodes, X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_nodes)
    model.fit(X_train, y_train)
    predicted = model.predict(X_test)
    mae = mean_absolute_error(y_test,predicted)
    return mae
```

```python
for max_node in [36, 39, 38]:
    print(max_node)
    print(get_mae(max_node, X_train, X_test, y_train, y_test))
    print('\n')
```
